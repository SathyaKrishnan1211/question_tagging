## CODEFORCES

```python
import requests
from bs4 import BeautifulSoup
```

```python
root = "https://codeforces.com/"
problemset_page = f"{root}problemset"
problemset_page
```

```python
result_problemset = requests.get(problemset_page)
result_problemset
```

```python
soup_cf_problemset = BeautifulSoup(result_problemset.text, "lxml")
soup_cf_problemset
```

```python
table_cf_problemset = soup_cf_problemset.find("table", class_="problems")
```

```python
all_qns = []
all_tags = []

for i in range(1, 83):

    page_i_link = f"{root}problemset/page/{i}"
    result_page = requests.get(page_i_link)
    soup_page = BeautifulSoup(result_page.text, "lxml")

    table_page_problemset = soup_page.find("table", class_="problems")
    div_page_problemset = table_page_problemset.find_all("div", style = "float: left;")
    problem_tags_page_problemset = table_cf_problemset.find_all("div", style = "float: right; font-size: 1.1rem; padding-top: 1px; text-align: right;")

    for j in range(len(div_page_problemset)):
        qn_link = div_page_problemset[j].find("a", href = True)["href"]
        qn_link = f"{root}{qn_link}"
        result_qn_link = requests.get(qn_link)
        soup_qn = BeautifulSoup(result_qn_link.text, "lxml")
        div_qn = soup_qn.find("div", class_ = "problem-statement")
        try:
            all_qns.append(div_qn.find_all("div")[10].text)
            tags = []
            for tag in problem_tags_page_problemset[j].find_all("a", class_ = "notice"):
                    tags.append(tag.text)
               
            all_tags.append(tags)
        except:
            pass

        

len(all_qns), len(all_tags)
```

```python
import pandas as pd
df_cf = pd.DataFrame(data = {"qns" : all_qns, "tags" : all_tags})
```

```python
df_cf.to_csv("codeforces.csv")
```

```python
from selenium import webdriver
import pandas as pd
import time
```

```python
root = "https://www.leetcode.com"
problemset = f"{root}/problemset/all"
all_qns = []
all_tags = []
links = []

driver = webdriver.Chrome("/home/sathya/Desktop/chromedriver")

for page_number in range(1,51):
    page = f"{problemset}/?page={page_number}"
    driver.get(page)
    time.sleep(5)
    div_qns = driver.find_elements_by_xpath("//div[@class='truncate overflow-hidden']")
    
    for div in div_qns:
        links.append(div.find_element_by_xpath("./a").get_attribute("href"))
    print("Len of links: ",len(links))
```

```python
len(links)
```

```python
all_qns = []
all_tags = []
```

```python
for link in links:
    driver.get(link)
    time.sleep(5)
    content = driver.find_element_by_tag_name("body")
    try:
        all_qns.append(content.find_element_by_xpath('//div[contains(@class,"question-content")]').text)
        temp_tags = []
        for tag in content.find_elements_by_xpath('//a[contains(@class,"topic-tag")]'):
            temp_tags.append(tag.get_attribute("href"))
        all_tags.append(temp_tags)
    except:
        pass
    print(len(all_qns), len(all_tags))
```

```python
import pandas as pd
df = pd.DataFrame(data={"qns" : all_qns,"tags" : all_tags})
df.to_csv("leetcode.csv")
```

```python
driver.quit()
```
